```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt

# Load the training data directly from the uploaded file
df_train = pd.read_csv('/content/train.csv')

# --- 1. Basic Information ---
print("First 5 Rows:")
print(df_train.head())
print("\nData Info:")
df_train.info()
print("\nDescriptive Statistics:")
print(df_train.describe())

# --- 2. Correlation Heatmap ---
# Select only numeric columns for correlation
numeric_cols = df_train.select_dtypes(include=['number'])
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Features')
plt.show()

# --- 3. Distribution of the Target Variable ---
fig = px.histogram(
    df_train,
    x='remaining_safe_storage_days',
    title='Distribution of Remaining Safe Storage Days',
    nbins=50
)
fig.show()
```
